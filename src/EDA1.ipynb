```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cleaned_race_data(1).csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='Stops', data=df)
plt.title('Distribution of Pit Stops in F1 Races')
plt.ylabel('Number of Pit Stops')
plt.show()
```

```python
columns_to_correlate = ['Race Position', 'Pos_grid', 'Laps', 'PTS_race', 'Stops',]
corr_matrix = df[columns_to_correlate].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix: Variables Affecting Race Position')
plt.show()
```
